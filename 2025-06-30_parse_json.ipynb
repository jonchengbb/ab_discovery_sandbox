```python
def max_residue_number(pdb_text: str) -> int:
    """
    Scan through a PDB-formatted string and return the maximum residue sequence number.
    """
    max_res = 0
    for line in pdb_text.splitlines():
        # Only consider coordinate records
        if line.startswith(("ATOM  ", "HETATM")) and len(line) >= 26:
            # Columns 2326 (0-based indices 22:26) hold the residue seq num
            res_str = line[22:26].strip()
            if res_str.isdigit():
                num = int(res_str)
                if num > max_res:
                    max_res = num
    return max_res


# Example usage:
# pdb_text = """\
# ATOM      1  N   MET A   1      11.104  13.207   9.100  1.00 20.00           N  
# ATOM      2  CA  MET A   1      12.560  13.327   9.350  1.00 20.00           C  
# ATOM    213  C   LYS A 100      23.456  17.890  11.234  1.00 20.00           C  
# HETATM  315  O   HOH A 250      19.876  15.234  14.789  1.00 20.00           O  
# """
print("Max residue number:", max_residue_number(pdb_text))
#  Max residue number: 250

```

```python
def analyze_multimer(pdb_text: str):
    """
    Given a PDB as a single string, report:
      - chains present
      - min/max residue number per chain
      - total unique residues per chain (ignoring insertion codes)
    """
    residue_info = {}
    for line in pdb_text.splitlines():
        if not (line.startswith("ATOM  ") or line.startswith("HETATM")):
            continue
        # Col 22 = chain ID, Cols 2326 = res seq num, Col 27 = insertion code
        chain_id = line[21]
        res_seq  = int(line[22:26].strip())
        ins_code = line[26]  # may be blank
        
        info = residue_info.setdefault(chain_id, {
            "min": res_seq, "max": res_seq, "unique": set()
        })
        info["min"] = min(info["min"], res_seq)
        info["max"] = max(info["max"], res_seq)
        # include insertion code if you want to distinguish e.g. 100A vs 100B
        info["unique"].add((res_seq, ins_code))
    
    # Print summary
    print("Chains found:", ", ".join(sorted(residue_info)))
    for chain, info in sorted(residue_info.items()):
        print(
            f"Chain {chain}: residues {info['min']}{info['max']}  "
            f"({len(info['unique'])} unique positions)"
        )


# # Example
# pdb_text = """\
# ATOM      1  N   MET A   1      11.104  13.207   9.100  1.00 20.00           N  
# ATOM      2  CA  MET A   1      12.560  13.327   9.350  1.00 20.00           C  
# ATOM    213  C   LYS A 100      23.456  17.890  11.234  1.00 20.00           C  
# ATOM      4  N   VAL B   1      15.000  12.000  10.000  1.00 20.00           N  
# ATOM      5  CA  VAL B   1      16.000  12.500  10.500  1.00 20.00           C  
# ATOM      6  C   VAL B  50      20.000  15.000  13.000  1.00 20.00           C  
# ATOM      7  N   GLY C   1      25.000  18.000  14.000  1.00 20.00           N  
# ATOM      8  CA  GLY C 200      30.000  20.000  15.000  1.00 20.00           C  
# """
analyze_multimer(pdb_text)
#  Chains found: A, B, C
#  Chain A: residues 1100  (2 unique positions)
#  Chain B: residues 150   (2 unique positions)
#  Chain C: residues 1200  (2 unique positions)
```

```python


heavy = atom_df[atom_df.element_symbol != "H"].copy()
heavy = heavy[heavy.residue_name.isin(AA_CODES)]

heavy = heavy[heavy.occupancy > 0]
heavy = heavy.drop_duplicates(
    ["chain_id","residue_number","insertion","atom_name","x_coord","y_coord","z_coord"],
)
# Identifier for each residue per chain
heavy["rid"] = list(zip(heavy.chain_id,
                        heavy.residue_number,
                        heavy.insertion))

ab_atoms = heavy[heavy.chain_id.isin(AB_CHAINS)].reset_index(drop=True)
ag_atoms = heavy[heavy.chain_id.isin(AG_CHAINS)].reset_index(drop=True)

delta_coords = (ab_atoms[["x_coord","y_coord","z_coord"]].to_numpy()[:, None, :] -
     ag_atoms[["x_coord","y_coord","z_coord"]].to_numpy()[None, :, :])

# m x n array of all heavy-atom distances
dist = np.linalg.norm(delta_coords, axis=-1) 

ab_atoms["min_d"] = dist.min(axis=1)

paratope_df = (
    ab_atoms                      # still has chain / number / AA name
      .groupby("rid", as_index=False)          # group all atoms in each residue
      .agg({
          "min_d": "min",                     # the number we need
          "chain_id": "first",                # keep useful metadata
          "residue_number": "first",
          "residue_name": "first"
      })
)
paratope_df['is_paratope'] = paratope_df['min_d'] <= 4.5

# paratope_df.to_csv("paratope_labels.csv", index=False)
```

```python
# 1build a mapping from every residue (rid) to its 0-based
#     position in the concatenated AlphaFold sequence.
rid2af_index = {}          # {(chain,resnum,icode): 0N-1}

# go through the heavy-atom DataFrame **once in file order**
running_index = 0
for rid in heavy.drop_duplicates("rid")["rid"]:
    rid2af_index[rid] = running_index
    running_index += 1

# 2add that integer to your paratope table
paratope_df["af_index"]  = paratope_df["rid"].map(rid2af_index)

# 3derive a chain-local index if you want
#     (handy when slicing VH vs VL blocks)
paratope_df["local_idx"] = (
        paratope_df["af_index"]
      - paratope_df["af_index"].where(paratope_df.chain_id == "A").min()   # VH starts at 0
)

# Now the earlier lambda makes sense
def min_cross_pae(row):
    if row.chain_id == "A":          # VH block
        return VH_vs_Ag[row.local_idx, :].min()
    else:                            # VL block
        return VL_vs_Ag[row.local_idx, :].min()
```

```python
min_pae_VH = VH_vs_Ag.min(axis=1)   # one number per VH residue
min_pae_VL = VL_vs_Ag.min(axis=1)

# attach to your paratope table
paratope_df["min_cross_PAE"] = paratope_df.apply(
    lambda r: min_pae_VH[r.local_idx] if r.chain_id=="A"
              else min_pae_VL[r.local_idx - len_VH],
    axis=1
)

# flag uncertain contacts
paratope_df["placement_ok"] = paratope_df.min_cross_PAE <= 10.0
```

```python
# order of residues in the AlphaFold FASTA (and in the PAE matrix)
chain_order = ["A", "B", "C"]          # VH  VL  antigen
len_dict = {                           # quick lookup table
    "A": len_VH,
    "B": len_VL,
    "C": len_Ag
}

# go through heavy-atom df once in file order
rid2af_index = {}
running = 0
for rid in heavy.drop_duplicates("rid")["rid"]:
    rid2af_index[rid] = running
    running += 1

paratope_df["af_index"] = paratope_df["rid"].map(rid2af_index)

# local index = position inside its own chain block
paratope_df["local_idx"] = (
    paratope_df.groupby("chain_id")["af_index"]
               .transform(lambda s: s - s.min())      # 0..len_chain-1
)

min_pae_VH = VH_vs_Ag.min(axis=1)      # shape (len_VH,)
min_pae_VL = VL_vs_Ag.min(axis=1)      # shape (len_VL,)

def cross_pae(row):
    if row.chain_id == "A":            # VH
        return min_pae_VH[int(row.local_idx)]
    else:                              # VL
        return min_pae_VL[int(row.local_idx)]

paratope_df["min_cross_PAE"] = paratope_df.apply(cross_pae, axis=1)
paratope_df["placement_ok"]  = paratope_df.min_cross_PAE <= 10.0
```
